To change the map background we can use tiles

```python
map = folium.Map(location = [20.5937,78.9629],tile = "CartoDB",zoom_start = 6)
map
```

To add Markers on map

```python
map.add_child(folium.Marker( location=[23.0225 , 72.5714] , popup = "Hello, I am Nisarg" , icon = folium.Icon(color = "blue") ))
map
```

```python
#More organised way to add map elements
fg = folium.FeatureGroup(name = "My Map")
fg.add_child(folium.Marker( location=[23.0225 , 72.5714] , popup = "Hello, I am Nisarg" , icon = folium.Icon(color = "blue") ))
map.add_child(fg)
```

To add multiple markers

```python
lst = [[23.1016470666409, 72.5401283200173],[23.11490966608841, 72.51429328292976],[12.939125644902827, 77.5841924303938],[19.077006468910504, 72.87018649106977],[17.391186180859563, 78.45330997581465]]
for coordinates in lst:
    fg.add_child(folium.Marker( location=coordinates , popup = "Hello, I am Nisarg" , icon = folium.Icon(color = "blue") ))
map.add_child(fg)    
```
